### 9. Find the top 25 movies in terms of average ratings for movies that have been rated more than 100 times - 10 marks

Hints : 

1. Find the count of ratings and average ratings for every movie.
2. Slice the movies which have ratings more than 100.
3. Sort values according to average rating such that movie which highest rating is on top.
4. Select top 25 movies.
5. You will have to use the .merge() function to get the movie titles.

Note: This question will need you to research about groupby and apply your findings. You can find more on groupby on https://realpython.com/pandas-groupby/.

```python
#Find the top 25 movies in terms of average ratings for movies that have been rated more than 100 times
df_movieData.isnull().sum()

df_userRatingData = pd.read_csv('Data.csv')
df_userRatingData.head()

# Count of Rating
df_userRatingData['rating'].value_counts()

# Average rating
df_userRatingData['rating'].mean()


# Slice the movies which have ratings more than 100.
df_userRatingData1 = df_userRatingData[df_userRatingData['rating'] > 1] 
df_userRatingData1

# Sort values according to average rating such that movie which highest rating is on top.
df1 = df_userRatingData1.sort_values('rating',ascending = False).groupby('movie id').head(2)

df1

# Select top 25 movies
top = df1.nlargest(25,'rating')
# print("Data_quantile(25%):",df1['rating']).quantile(0.75)


# You will have to use the .merge() function to get the movie titles.
df2 = pd.merge(top, df_movieData, on='movie id', how='left')
df2.head()

df2['movie title']
```
